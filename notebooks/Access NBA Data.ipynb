## Access NBA through API

```python
BASE_URL = "https://dsci511-nba-api.herokuapp.com/"
```

```python
requests.get(BASE_URL).json()
```

### Get all NBA Teams

```python
response = requests.get(f"{BASE_URL}teams")
response
```

```python
response.json()
```

```python
pd.DataFrame(response.json())
```

### Get Team Roster by Team ID

```python
team_id = 1610612755
response = requests.get(f"{BASE_URL}roster/{team_id}")
response
```

```python
response.json()
```

```python
pd.DataFrame(response.json())
```

### Get Player Information by Player ID

```python
player_id = 1630178
response = requests.get(f"{BASE_URL}player/{player_id}")
response
```

```python
response.json()
```

```python
pd.DataFrame([response.json()])
```

### Get Quick Player Stats by Player ID

```python
player_id = 1630178
response = requests.get(f"{BASE_URL}player/quickstats/{player_id}")
response
```

```python
pd.DataFrame([response.json()])
```

```python
# Can do a inner join between these two pieces of data
player_id = 1630178
player_response = requests.get(f"{BASE_URL}player/{player_id}")
quickstat_response = requests.get(f"{BASE_URL}player/quickstats/{player_id}")

player_df = pd.DataFrame([player_response.json()])
quickstat_df = pd.DataFrame([quickstat_response.json()])

player_df.merge(quickstat_df, on="player_id")
```
